## Gathering logs across runs of Algo/Game pairs

```python
game = 'ZaxxonNoFrameskip-v4'
algo = 'ga'

from atari_zoo.log import gather_logs_across_runs, gather_logs_across_algos
```

```python
df = gather_logs_across_runs(algo,game,range(1,4),local=False)
```

### Plotting with seaborn

```python
import seaborn as sns
```

```python
sns.lineplot(x="frames", y="score",
             style="run", hue='algo',
             data=df)
```
